```python
import cv2
from ultralytics import YOLO
model_detect = YOLO("models/Yolo_face_detection.pt")
```

## Пример сообщения из ultralytics.YOLO

```python
cap = cv2.VideoCapture(0)


status, img = cap.read()

try:
    results = model_detect(img)
    
    print(results)
    print(results[0].boxes)

except Exception as e:
    print(f'Error: {e}')
    
cap.release()
cv2.destroyAllWindows()
```

## Пример работы с изначальным модулем YOLO

```python
cap = cv2.VideoCapture(0)
while True:
    status, img = cap.read()
    if not status:
        print('error status img')
        break


    try:
        results = model_detect(img)
        
        if results and results[0].boxes is not None:
            boxes = results[0].boxes
            if len(boxes) > 0:
                box = boxes[0]
                x1, y1, x2, y2 = map(int, box.xyxy[0])
                face = img[y1:y2, x1:x2]
                
                # cv2.putText(img, text, (x1, y1 - 10),
                            # cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 0, 255), 2)
                cv2.rectangle(img, (x1, y1), (x2, y2), (0, 255, 0), 2)

    except Exception as e:
        print(f'Error: {e}')
        continue

    
    
    cv2.imshow('Image', img)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        print('exit')
        break


cap.release()
cv2.destroyAllWindows()
```

## Пример работы с "UTILS.YOLO"

Эта утилита нужна для более удобной работы с фотографиями, в нем можно 1 раз прописать все необходимые вещи (например порог срабатывания, минимальный размер, встроенные ID и т.п.)

```python
from utils.Yolo import Yolo as Yolo
import time
import cv2

yolo = Yolo()


# while True:
#     package = next(yolo)

for package in yolo:


    if package["num_faces"] != 0:
        for i, face in enumerate(package["faces"]):
            face_img = face["face_img"]
            
            cv2.imshow(f"face{i}", face_img)
    
    time.sleep(1)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
    
yolo.close()
cv2.destroyAllWindows()
```
